```python
import pymysql
import cryptography
import os 
```

```python
## To get the actual Password
pass_word = os.getenv("MYSQL_PASSWORD")
```

```python
# To establish a connection between Python and MySQL
db = pymysql.connect(
    host = "localhost",
    port = 3306,
    user = "root", 
    password = pass_word
)
cursor = db.cursor()
if cursor:
    print("Connection was successful..")
```

```python
## To display all the database in MySQL
cursor.execute("Show databases")
for i in cursor:
    print(i)
```

```python
# To create a Database in Python 
cursor.execute("CREATE DATABASE EC_MFB")
cursor.execute("Show databases")
for i in cursor:
    print(i)
```

```python
# To create a table 
cursor.execute("use ec_mfb")
cursor.execute(""" create table loan_rd(
Loan_id int primary key not null, 
Full_name Varchar(50) not null, 
BVN int not null, 
Phone_no Varchar(20) not null,
Gender enum("M", "F") not null,
Occupation VARCHAR(25) not null, 
Estimated_income int not null,
rd_created Timestamp DEFAULT CURRENT_TIMESTAMP)""")
if cursor:
    print("Table was created..")
```

```python
cursor.execute("describe loan_rd")
for i in cursor:
    print(i)
```

```python
# To build a system that can prompt the user for certain information 
cursor.execute("use ec_mfb")
def register():
    print("="*40)
    print("Kindly answer the following questions")
    Full_name = input("Enter your Full name: ")
    BVN = int(input("Enter your BVN: "))
    Phone_no = input("Enter your Phone number: ")
    Gender= input("Enter your Gender: ")
    Occupation = input("What do you do for a living: ")
    Estimated_income = int(input("Enter your estimate income: "))

    import string as st 
    import random 
    num = st.digits
    loan_id = int(''.join(random.choice(num) for _ in range(5)))
    query = "insert into loan_rd() values(%s,%s,%s,%s,%s,%s,%s, DEFAULT)"
    cursor.execute(query, (loan_id, Full_name, BVN, Phone_no, Gender, Occupation, Estimated_income))
    db.commit()
    print(f'Registration was successful, Your Loan_id is {loan_id}.')

def find():
    print("="*40)
    loan_id = int(input("Enter your loan_id: "))
    query = f"SELECT * FROM loan_rd WHERE loan_id = {loan_id}"
    cursor.execute(query)
    result = cursor.fetchall()
    if len(result) == 0:
        print("Records not found\n")
    else:
        for i in result:
            print(i)
def update():
    print("="*40)
    loan_id = int(input("Enter your loan_id: "))
    Full_name = input("Enter your Full name: ")
    BVN = int(input("Enter your BVN: "))
    Phone_no = input("Enter your Phone number: ")
    Gender= input("Enter your Gender: ")
    Occupation = input("What do you do for a living: ")
    Estimated_income = int(input("Enter your estimate income: "))

    query = "update loan_rd SET Full_name = %s, BVN = %s, Phone_no = %s, Gender = %s, Occupation = %s, Estimated_income = %s WHERE loan_id = %s"
    values = (Full_name, BVN, Phone_no, Gender, Occupation, Estimated_income, loan_id)
    cursor.execute(query, values)
    db.commit()
    print("Records have been updated.\n")

def apply():
    print("="*40)
    loan_id = int(input("Enter your loan_id: "))
    Full_name = input("Enter your Full name: ")
    Loan_amount = int(input("Enter loan amount: "))
    Estimated_income = int(input("Enter your estimate income: "))
    if Estimated_income < 200000:
        print("Sorry, you are not eligble to get loan\n")
    else:
        query = "insert into cus_rd(loan_id, Full_name, Loan_amount, Estimated_income) value(%s, %s, %s, %s)"
        values = (loan_id, Full_name, Loan_amount, Estimated_income)
        cursor.execute(query, values)
        print("Loan request was accepted\n")
def main():
    while True:
        print("="*40)
        print("<<<< Welcome Earlycode Loan system >>>")
        print("1 <<<< Registration")
        print("2 <<<< To check records")
        print("3 <<<< To update your record ")
        print("4 <<<< Apply for Loan")
        print("5 <<<< Logout")
        option = input("Kindly make a selection:")   
        if option == "1":
            register()
        elif option == "2":
            find()
        elif option == "3":
            update()
        elif option == "4":
            apply()
        elif option == "5":
            print("Thank for patronage..\n")
            break
        else:
            print("Invaild selection")
```
